```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
test = pd.read_html(url, skiprows = [1])
print(type(test))
print(len(test))
```

```python
#Determine the Data type of the first item in a list
type(test[1])
#Accessing the second item in a list
df = test[1]
#df.set_index("Rank", inplace = True)
print("df shape", df.shape)
print(df.info())
df.head()
```

```python
#Drop cols with NAN
df.drop(columns = "Population", inplace = True)
df.info()
```

```python
#Set Date to DatetimeIndex
df["Date"] = pd.to_datetime(df["Date"].values.flatten())
#View the first five cols
df.head()
```

```python
import requests
```

```python
#Make an http request, assign it to the variable html
html = requests.get(url).text
print(type(html))
```

```python
#use dir to access all the attributes of html
dir(html)
```

```python
# parse the html in a BeautifulSoup, assign it to the variable soup
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify())
```

```python
#Use dir to find all the attributes of soup
dir(soup)
```

```python
my_table = soup.find("table")
print(my_table)
```

```python
dir(my_table)
my_table.find_all("tbody")
```

```python
for body in my_table.find_all("tbody"):
    print(body.text)
```

**WEB SCRAPING INDEED.COM**

```python
page_start = 10
url = f"https://www.indeed.com/jobs?q=data+scientist&l=Los+Angeles%2C+CA&radius=35&start={page_start}"
url
```

```python
results = []
for page in range(0, 50, 10):
    url = f"https://www.indeed.com/jobs?q=data+scientist&l=Los+Angeles%2C+CA&radius=35&start={page_start}"
    
    #make an http request
    html = requests.get(url).text
    #parse html using Beautiful Soup
    soup = BeautifulSoup(html, "html.parser")
    #append to list
    results.append(soup)
print(results[0])
```
